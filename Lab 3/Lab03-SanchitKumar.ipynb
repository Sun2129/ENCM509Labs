# Introduction

This lab is an introduction into biometric data analysis and how to use such data in order to match and analyze Genuine and Impostor signatures. The same method as in Lab 2 was used to acquire the data required for the exercises in this lab. On top of this, this lab introduces the idea of classifiers and how to train models in order to classify future Genuine and Impostor data. The Gaussian Mixture Model will be used to calculate Genuine and Impostor scores in order to further analyze the data through statistical and graphical analysis. Through the Genuine and Impostor scores, it becomes possible to calculate critical regions and FAR and FRR values. At the same time, such scores will be represented graphically in order to gain a deeper understanding of biometric data analysis. 

### Exercise 1

The following 12 graphs are of the Normal Distributions of Genuine and Impostor Scores calculated using a Gaussian Mixture Model that was trained using various amounts of training set data (various amounts of Genuine signatures) and various amounts of Gaussian components. The Genuine and Impostor signatures are of the word "biometrics" found in the subdirectories "genuine" and "impostor" of the folder "biometrics". 4 sets/combinations of training set data and Gaussian components were used to create 4 different plots of Normal Distribution of Genuine and Impostor scores. The first triplet of graphs uses 20 genuine signatures as the training set data and 20 Gaussian components. The second triplet of graphs uses 10 genuine signatures as the training set data and 10 Gaussian components. The third triplet of graphs uses 20 genuine signatures as the training set data and 5 Gaussian components. The fourth triplet of graphs uses 25 genuine signatures and 40 Gaussian components. All of the impostor curves below use 30 impostor signatures as the data.

As can be seen from the results, lowering both the Training Set Data and the Gaussian components leads to overall higher mean Genuine and Impostor scores. On the other hand, lowering simply the the number of Gaussian components leads to overall lower mean Impostor scores. Finally, lowering the number of Training Set Data and increasing the number of Gaussian components leads to overall higher mean Genuine and Impostor scores. 
The standard deviations also seem to follow these results. Lowering the Training Set Data and Gaussian components leads to higher standard deviations for the genuine scores and slightly lower standard deviations for the impostor scores (although this may change based on the run). Lowering only the number of Gaussian components leads to slightly lower standard deviations for the genuine scores and significantly lower standard deviations for the impostor scores. Lastly, lowering the number of Training Set Data and increasing the number of Gaussian components leads to higher standard deviations for the genuine scores and significantly higher standard deviations for the impostor scores. As a result of these changes, the graphs and curves produced seem to shift right or left depending on the selected amount of Training Set Data and Gaussian components.
Overall, varying the Training Set Data and Gaussian components has great effects on the scores produced for both Genuine and Impostor scores.

When plotting these 12 graphs, I found that it was always better to plot the Genuine and Impostor graphs side by side rather than plotting them on top of each other. The information can be displayed better when it is two separate graphs rather than one singular graph. All points of information and the distribution itself are shown a lot more nicely when separated into two graphs. This is especially true when the overall y values of both graphs differ by a considerable amount. For example, in the graph in section 1.a where the curves are on the same graph, the overall y values of the green curve are significantly higher than those of the red curve. Because of this, it becomes significantly harder to view the red curve properly (along with points of interest). As they also use the same x range, and because the red curve requires a longer x range due to its data, the green curve also becomes harder to view as it requires a much smaller x range (the green graph gets squished). These same problems can be seen in sections 1.a, 1.b, 1.c, and 1.d when comparing the graphs that are side by side and the graph that contains two curves. On the other hand, with the graphs in section 1.a that are side by side, it is significantly easier to view the curves properly as they are side by side and use different x ranges and y values. Points of interest and the curves overall are easier to see if each curve is represented separately. The only slight problem is that when interpreting these curves side by side, it is important to keep in mind the difference between the x and y ranges of both graphs.

#### Exercise 1.a

This first graph uses 20 genuine signatures as training set data and 20 Gaussian components. As can be seen from the Genuine and Impostor graphs, the Genuine scores are more likely to be around the value of -17 while the Impostor scores are more likely to be around -300. While most values of Impostor scores do not overlap with Genuine scores, there is a section of the Impostor graph that does overlap the graph of the Genuine scores. The mean of the Genuine scores was around -17, while the mean of the Impostor scores was -300. On the other hand, the standard deviation of the Genuine scores was around 0.38 and the standard deviation of the Impostor scores was around 173. 

The graph shown below the two genuine and impostor graphs is another graph which has the same properties as the first two graphs. The only difference is that the genuine and impostor curves have been plotted on the same graph rather than being separated and shown side by side. The effects of this have been detailed in the overall Exercise 1 section.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import pandas as pd
from sklearn.mixture import GaussianMixture
from scipy.stats import norm

# data folder:
# 'signature_samples/biometrics/'
# 'signature_samples/calgary/'
# 'signature_samples/hello/'

# this is the place where the folders "genuine" and "impostor" are
# change it to point to your dataset.
base_dir = 'signature_samples/biometrics/'
# note that if your dataset folder is in the same folder as the notebook, 
# you don't need the "../" only "./"
#base_dir = './signature_samples/biometrics/'

# "genuine" and "impostor" sub-directories
dir_authentic = base_dir + 'genuine/'
dir_impostor = base_dir + 'impostor/'

# amount of signatures to be loaded
total_signatures = 30
numForgedSigs = 30

# lists to store the loaded signatures
auth_lst = []
imp_lst = []

# loading 'total_signatures' signatures from both "authentic" and "impostor" 
for i in range(total_signatures):
    # columns: [X, Y, Pressure, Time]
    csv_auth = pd.read_csv(dir_authentic + '%d.csv' % (i+1))
    csv_imp = pd.read_csv(dir_impostor + '%d.csv' % (i+1))

    auth_lst.append(csv_auth.values)
    imp_lst.append(csv_imp.values)

# number of signature samples to use for training the models
NUMTRAININGPATTERNS = 20

# number of Gaussian mixture components in the GMM
NUMGCOMPONENTS = 20

# usually we don't know the final size of train_set because
# each signature has different number of points. 
# to later on use 'concatenate', is necessary to take the first sample outside the loop
train_set = np.reshape(auth_lst[0], (-1,4))

# this for-loop put all the first "NUMTRAININGPATTERNS" in a huge matrix
# this matrix, stored in the varaible "train_set" will be used for the training
for i in range(1,NUMTRAININGPATTERNS):
    # taking from authentic/genuine
    sample = np.reshape(auth_lst[i], (-1,4))
    train_set = np.concatenate((train_set, sample), axis=0)

# Training the GaussianMixture model from Scikit-Learn library
# the .fit(...) will perform the training step using the "train_set" 
gmm = GaussianMixture(n_components=NUMGCOMPONENTS).fit(train_set)

# to store all the scores calculated
aScores = []

# going through the remaining signatures after removing the first 20 for training
for i in range(NUMTRAININGPATTERNS, total_signatures):
    # taking from authentic/genuine
    sample =  np.reshape(auth_lst[i], (-1,4))
    # compute the per-sample average log-likelihood of the given data.
    score = gmm.score(sample)
    print('%d: %.4f' % (i, score))
    # add the score calculated to the list aScores
    aScores.append(score)

fScores = []
for i in range(numForgedSigs):
    # taking from the impostor
    sample =  np.reshape(imp_lst[i], (-1,4))
    score = gmm.score(sample)
    print('%d: %.4f' % (i, score))
    fScores.append(score)

# authentic set
aMu = np.mean(aScores)
aStd = np.std(aScores)

# impostor set
fMu = np.mean(fScores)
fStd = np.std(fScores)

print('Average of authentic scores: %.4f' % (aMu))
print('Average of forgery scores: %.4f' % (fMu))
print('Standard Deviation of authentic scores: %.4f' % (aStd))
print('Standard Deviation of forgery scores: %.4f' % (fStd))

plt.figure(figsize=(20,4))

# Authentic set
x = np.arange(-20, -15, 0.01)
aProb = norm.pdf(x, loc=aMu, scale=aStd)

plt.subplot(1,2,1)
plt.plot(x, aProb, 'g')
plt.plot(aScores, np.zeros(len(aScores)), 'gs', lw=2, label='Authentic')
plt.legend();
plt.title('Normal Distribution of Genuine Signatures with Training set: {} - Gaussian components: {}'.format(NUMTRAININGPATTERNS, NUMGCOMPONENTS));


# Forged (impostor) set
x = np.arange(-800, 50, 0.01)
fProb = norm.pdf(x, loc=fMu, scale=fStd)

plt.subplot(1,2,2)
plt.plot(x, fProb, 'r')
plt.plot(fScores, np.zeros(len(fScores)), 'r^', lw=2, label='Forgery')
plt.legend();
plt.title('Normal Distribution of Impostor Signatures with Training set: {} - Gaussian components: {}'.format(NUMTRAININGPATTERNS, NUMGCOMPONENTS));
plt.xlabel('')
```

The level of significance is $\alpha = 0.05$. Therefore, $z_{\frac{\alpha}{2}} = z_{\frac{0.05}{2}} = z_{0.025} = 1.96$.

In Exercise 1.a, the mean calculated for the Genuine scores is $\overline{x} = -17.1279$ and the standard deviation is $\sigma = 0.3839$. The mean calculated for the Impostor scores is $\overline{x} = -316.4347$ and the standard deviation is $\sigma = 172.6207$. In this exercise, 20 Genuine signatures were used as Training Data Set and so only 10 Genuine signatures were actually used on the completed model.

#### Critical Region Calculation:
The critical regions are: $\overline{x} - z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}} < \mu < \overline{x} + z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}}$.

So, $-17.1279 - 1.96\frac{0.3839}{\sqrt{10}} < \mu < -17.1279 + 1.96\frac{0.3839}{\sqrt{10}}$.
Finally, the critical regions are: $-17.366 < \mu < -16.890$. 

My hypothesis will be that $H_{0}$: $\mu = -17.0$ and $\sigma = 0.38$.

In our sample of 10, there are 5 scores that fall outside of the critical region (-16.8582, -16.7444, -17.8539, -17.4023, and -16.3873).

So, $FRR(FNMR) = 5/10 = 0.5$, which is a somewhat large error. 

#### FRR Calculation:
Changing the critical region from: $-17.366 < \mu < -16.890$ to $-17.4 < \mu < -16.7$.

So, the new critical values are: $z_{1} = \frac{\overline{x_{1}} - \mu}{\frac{\sigma}{\sqrt{n}}}$ and $z_{2} = \frac{\overline{x_{2}} - \mu}{\frac{\sigma}{\sqrt{n}}}$.

So, $z_{1} = \frac{-17.4 - (-17.0)}{\frac{0.38}{\sqrt{10}}}$ and $z_{2} = \frac{-16.7 - (-17.0)}{\frac{0.38}{\sqrt{10}}}$.

Finally, $z_{1} = -3.33$ and $z_{2} = 2.50$.

Therefore, $FRR = P(Z < -3.33) + P(Z > 2.50) = P(Z < -3.33) + (1 - P(Z < 2.50)) =  0.0004 + (1 - 0.9938) = 0.0066$.

#### FAR Calculation
My hypothesis will be that $H_{1}$: $\mu = -316.4347$ and $\sigma = 172.6207$. 

So, $z_{1} = \frac{\overline{x_{1}} - \mu}{\frac{\sigma}{\sqrt{n}}}$ and $z_{2} = \frac{\overline{x_{2}} - \mu}{\frac{\sigma}{\sqrt{n}}}$.

With the critical region of $-17.366 < \mu < -16.890$, $z_{1} = \frac{-17.366 - (-316.4347)}{\frac{172.6207}{\sqrt{30}}}$ and $z_{2} = \frac{-16.890 - (-316.4347)}{\frac{172.6207}{\sqrt{30}}}$.

Finally, $z_{1} = 9.49$ and $z_{2} = 9.50$.

Finally, $FAR = P(9.49 \le Z \le 9.50) = P(Z \le 9.50) - P(Z \le 9.49) = 1 - 1 = 0$.

Thus, the evaluated FRR is 0.0066 or 0.66% while the evaluted FAR is 0 or 0%.
