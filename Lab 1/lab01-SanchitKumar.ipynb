### Exercise 2:

The 3 plots/graphs below are graphs of the function $f(x) = sin(x) + \frac{cos(2*x)}{2}\ + \frac{sin(3*x)}{3}\ + \frac{cos(4*x)}{4}$ where $x \in (-6, 6)$ using different plotting styles. 

How to use plot and include different arguments for stylization was taken from https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html and different marker types was taken from https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers .

#### Exercise2.a: 

This first graphs plots the function mentioned above with the given interval. It also modifies the stylization of the graph so that the graph marker is "o", the linestyle is "--", and so that the color is "red". This was done by adding the arguments "marker = 'o'", "linestyle = '--'", and "color = 'red'" when calling the plot function from matplotlib.pyplot. As can be seen in the output, a graph with red dashed lines and circular markers is created when the code is ran.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

# Set the range/interval of x (-6, 6)
x = np.linspace(-6, 6)

# Create the function y = f(x) = sin(x) + cos(2*x)/2 + sin(3*x)/3 + cos(4*x)/4
y = np.sin(x) + np.cos(2*x)/2 + np.sin(3*x)/3 + np.cos(4*x)/4

# Plot the function with marker set to 'o', linestyle set to '--', and color set to red
plt.plot(x, y, marker = 'o', linestyle = '--', color = 'red')
# Set the title of the plot
plt.title('$f(x) = sin(x) + cos(2*x)/2 + sin(3*x)/3 + cos(4*x)/4$')
# Set the label of the x axis
plt.xlabel('x')
# Set the label of the y axis
plt.ylabel('$sin(x) + cos(2*x)/2 + sin(3*x)/3 + cos(4*x)/4$')
```

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

# Percentage of male and female
gender = [55 , 45]
# Labels for each part of the pie chart
gender_labels = ['Male', 'Female']

# Generate the pie chart with the appropriate labels, percentages shown, shadow enabled, and the pie slices somewhat separated
plt.pie(gender, labels = gender_labels, autopct = '%1.0f%%', shadow = 'True', explode = (0.03, 0.03))
# Title the pie chart and bold it
plt.title('Gender', fontweight = 'bold')
```

The following graph is a Normal Distribution of the age of the passengers and crew aboard the Diamond Princess Cruise Ship. The way to create the functions below in Markdown required these documentation: https://www.evanlray.com/stat242_f2019/resources/R/MathinRmd.html and https://math.meta.stackexchange.com/questions/22443/how-to-markup-expression-using-sigma-notation-for-sum-in-mathjax.

The formula for normal distribution, from the file LecWeek1-PostLab1+Ch3-1-ENCM509-W23 on D2L, is $N_{\mu,\sigma}(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$. This formula is not used in the code however, as we call the probability distribution function from norm called norm.pdf (as shown in the LecWeek1-PostLab1+Ch3-1-ENCM509-W23 on D2L).

The formula for mean, from the file LecWeek1-PostLab1+Ch3-1-ENCM509-W23 on D2L, is $\mu = \frac{1}{N} \sum\limits_{n=1}^N x_n$. This formula is not used in the code however, as we call the mean function in numpy (how to use numpy.mean was taken from https://numpy.org/doc/stable/reference/generated/numpy.mean.html).

The formula for standard deviation, from the file LecWeek1-PostLab1+Ch3-1-ENCM509-W23 on D2L, is $\sigma^2 = \frac{1}{N-1} \sum\limits_{n=1}^N (x_n-\mu)^2$. This formula is not used in the code however, as we call the standard deviation function in numpy (how to use numpy.std was taken from https://numpy.org/doc/stable/reference/generated/numpy.std.html).

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Counting of people at each age interval shown in age labels
# Give each pre-existing element of the age vector an index
age = [[0, 16], [1, 23], [2, 347] , [3, 428] , [4, 334] , [5, 398] , [6, 923] , [7, 1015] , [8, 216] , [9, 11]]
age_labels = ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-89','90-99']

# Initialize a new array to hold the amont of each index
index_list = []

# Add each index in the age array g amount of times where g is the number that has been indexed
# For each array inside of the age vector...
for x in age:
    # Create variable i and set it to 0
    i = 0
    
    # While i is less than the element in the second index of the array (like 16 or 23)...
    while i < x[1]:
        # Add the element in the first index of the array (this is the indexes that we added previously to the age vector)
        index_list.append(x[0])
        # Add 1 to i
        i += 1

# Get the mean of the new array
mean = np.mean(index_list)

# Get the standard deviation fo the new array
std = np.std(index_list)

# Create a range starting from the min element in index_list (0) to the max elements in index_list (9) with a step of 0.01 to smooth the created plot
index_list = np.arange(min(index_list), max(index_list), 0.01)

# Create the normal distribution plot by using the probability density function (norm.pdf)
plt.plot(index_list, norm.pdf(index_list, mean, std))

# Title the graph
plt.title('Normal Distribution of the Age of the Population on the Diamond Princess Cruise Ship')
# Title the x axis
plt.xlabel('Age')
# Title the y axis
plt.ylabel('pdf(Age)')
```
