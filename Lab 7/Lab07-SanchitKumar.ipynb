### Exercise 1

In this exercise, the percentage of data used for training and testing was varied (while all other variables were kept constant) in order to determine the effect on classifier accuracy. The data used in this exercise was taken from the "data" folder that was created after extracting the "data" zip file on D2L. There are two code snippets present: the first one evaluates the classifier accuracy when the variable test_size is set to 0.40 (or 40% of the data in the "data" folder is used for testing while 60% of the data is used for training the model) and the second one evaluates classifier accuracy when test_size is set to 0.20 (or 20% of the data in the "data" folder is used for testing and 80% of the data is used for training the model). 

The classifier accuracies produced were as follows:

| Percentage of Training Data | Percentage of Test Data | Classifier Accuracy|
|----------------|----------------|----------------|
|60%|40%|0.75|
|80%|20%|1.00|

Through running these tests, I found that, rather than being a concrete value, the classifier accuracy on the same code would fluctuate. I ran each code snippet around 10 times in order to try to gain an average classifier accuracy for each Percentage of Training Data and Percentage of Test Data test. So, rather than the classifier accuracy just being 0.75 when the test size is 40% of the data in the data folder, it would fluctuate between 0.625-1.00. On the other hand, the classifier accuracy would fluctuate between 0.75-1.00 when the test size is 20% of the data. Based on these values, it looks like increasing the number of data for the model to train on increases its resulting model and prediction accuracy (as the classifier accuracy range overall increased when we decreased the percentage of data used for testing). Consequently, if we were to run a test where we increased the percentage of data used for testing (which would result in decreasing the percentage of data used for training), then we would most likely see that the classifier accuracy range overall would decrease (although we would need to perform more tests to confirm this). A balance needs to be struck between the percentage of data used for tests and the percentage of data used for training as using too little data for training would provide us with an inaccurate model, while using too much data for training would mean that we would have little to no data to actually test on and obtain results. Overall, the amount of data used for training a model is incredibly important in order to create an accurate model and in order to be able to correctly classify test data.

```python
import numpy as np
import pandas as pd

#import tensorflow.keras as keras
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from os import listdir
from os.path import isfile, join, dirname, abspath
from sklearn.metrics import confusion_matrix

from utils_cm import plot_confusion_matrix

def build_model():
    model = models.Sequential()    
    # an input layer that expects:
    # 1 or more samples, NUMBER_TIMESTEPS time steps and NUMBER_FEATURES features.
    
    # 1st LSTM layer
    model.add(layers.LSTM(256, return_sequences=True, input_shape=(NUMBER_TIMESTEPS, NUMBER_FEATURES)) )
    
    # 2nd LSTM layer
    model.add(layers.LSTM(256, input_shape=(NUMBER_TIMESTEPS, NUMBER_FEATURES)) )
    
    # Hidden fully connected layers of the neural network
    model.add(layers.Dense(512,activation='relu'))  
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(256,activation='relu'))    
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(512,activation='relu'))
    
    # Classification layer of the neural network
    model.add(layers.Dense(NUMBER_OUTPUTS, activation='softmax'))
    
    opt = Adam(learning_rate=0.002)
    model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    
    # this shows the network structure 
    model.summary()
    
    return model

# the length of records (if shorter, we need to add some zero rows)
NUMBER_TIMESTEPS = 100
# the number of features (from the data)
NUMBER_FEATURES = 202
# the number of classes/gestures
NUMBER_OUTPUTS = 2
# you can encode more than 1 but for this example we have binary output (circle/swipe)

# the directory where your data is
mypath = './data'

# creating a list with all the filenames
datafiles = [f for f in listdir('data') if isfile(join(mypath, f))]

# choose data we need
columns = ['handPalmPosition_X','handPalmPosition_Y','handPalmPosition_Z',
          'pitch', 'roll', 'yaw', 'GestureTypeCircle', 'GestureTypeSwipe',
          'wristPosition_X', 'wristPosition_Y','wristPosition_Z',
          'elbowPosition_X', 'elbowPosition_Y', 'elbowPosition_Z']

finger_names = ['Thumb', 'Index', 'Middle', 'Ring', 'Pinky']
bone_names = ['Metacarpal', 'Proximal', 'Intermediate', 'Distal']
    
for finger in finger_names:
    columns.append(finger + 'Length')
    columns.append(finger + 'Width')

for finger in finger_names:
    for bone in bone_names:
        columns.append(finger + bone + 'Start_X')
        columns.append(finger + bone + 'Start_Y')
        columns.append(finger + bone + 'Start_Z')
        columns.append(finger + bone + 'End_X')
        columns.append(finger + bone + 'End_Y')
        columns.append(finger + bone + 'End_Z')
        columns.append(finger + bone + 'Direction_X') 
        columns.append(finger + bone + 'Direction_Y') 
        columns.append(finger + bone + 'Direction_Z')   
        
# Features
x = []
# Labels
y = []

for sample in datafiles:
    relative_path = 'data\\' + sample
    tmp = pd.read_csv(relative_path, usecols=columns)
    
    # Normalize the sample size: LSTM requires all inputs of the same size!
    print('{}\nsize raw = {}'.format(relative_path,tmp.shape))
    while tmp.shape[0] < NUMBER_TIMESTEPS:
        tmp = tmp.append(pd.Series(0, index=tmp.columns), ignore_index=True)

    if tmp.shape[0] > NUMBER_TIMESTEPS:
        tmp = tmp.head(100)
    print('size normalized = ',tmp.shape)
    
    tmp_x = tmp[[column for column in list(tmp.columns)
                 if column != 'GestureTypeCircle' 
                 and column != 'GestureTypeSwipe']]
    tmp_y = tmp[['GestureTypeCircle', 'GestureTypeSwipe']]
        
    x.append(tmp_x)
    y.append(tmp_y)

np.array(y[0].loc[0])

# Each sample requires labels of [1,NUMBER_OUTPUTS] size (not a list)
y_new = list()
for cur_label in y:
    tmp = np.array(cur_label.loc[0])
    y_new.append(tmp)
y = np.array(y_new)

# Set a percentage of test set fraction
test_size = 0.40 # 40%

# Divide data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=test_size, shuffle=True)
len_train = len(X_train)
len_test = len(X_test)

print ('Number of train samples = {}\nNumber of test samples = {}'.format(len_train, len_test))
print ('There is ',type(X_train),' of ',type(X_train[0]))

# Turn list(DataFrame) into numpy.ndarray with [len_train, NUMBER_TIMESTEPS, NUMBER_FEATURES]
X_train = np.array(X_train)
X_test = np.array(X_test)
print('The list was turned into <numpy.ndarray>')

# creating the model
model = build_model()

# training the model
model.fit(X_train, y_train, epochs=10, batch_size=len_train)

# testing the model
test_loss, test_acc = model.evaluate(X_test, y_test, batch_size=len_train)

y_pred = model.predict(X_test)

matches = (y_pred == y_test)
print('Total of matches: %d' % (matches.sum()))

match_rate = matches.sum() / float(len(matches))
print('Match rate: %.2f' % (match_rate))

print('Classifier Accuracy: %.2f' % test_acc)

y_pred_rounded = y_pred.round(decimals=0, out=None)

y_pred_cm = np.argmax(y_pred, axis = 1)
y_test_cm = np.argmax(y_test, axis = 1)

# Generate the confusion matrix
confusionMatrix = confusion_matrix(y_test_cm, y_pred_cm)

plot_confusion_matrix(cm=confusionMatrix,
                      target_names = [i for i in range(1, 3)])
```
