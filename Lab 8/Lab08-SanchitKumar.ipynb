### Part 1
First, we are trying to find the probability that a person has Influenza given that the person also has a Fever. Using Bayes' rule: 
$P(Influenza = T | Fever = T) = \frac{P(Fever = T | Influenza = T) * P(Influenza = T)}{P(Fever = T)}$

The top of the equation can be reduced as such: $P(Fever = T | Influenza = T) * P(Influenza = T) = P(Fever = T, Influenza = T).$ 

Then, based off of Total Probability, $P(Fever = T, Influenza = T) = P(W, C, I = T, F = T) = $

$\sum_{i}\sum_{j} P(W = i) * P(C = j | W = i) * P(I = T | W = i) * P(F = T | C = j, I = T) =$
$0.5 * 0.8 * 0.5 * 0.9 +
0.5 * 0.2 * 0.5 * 0.99 +
0.5 * 0.2 * 0.9 * 0.9 +
0.5 * 0.8 * 0.9 * 0.99 = 0.6669.$

Also, $P(Fever = T) = \sum_{i}\sum_{j}\sum_{k}P(W = i) * P(C = j | W = i) * P(I = k | W = i) * P(F = T | C = j, I = k) =$
$0.5 * 0.8 * 0.5 * 0.0 +
0.5 * 0.8 * 0.5 * 0.9 +
0.5 * 0.2 * 0.5 * 0.9 +
0.5 * 0.2 * 0.5 * 0.99 +
0.5 * 0.2 * 0.1 * 0.0 +
0.5 * 0.2 * 0.9 * 0.9 +
0.5 * 0.8 * 0.1 * 0.9 +
0.5 * 0.8 * 0.9 * 0.99 = 0.7479.$

Finally, $P(Influenza = T | Fever = T) = \frac{P(Fever = T | Influenza = T) * P(Influenza = T)}{P(Fever = T)} = \frac{P(Fever = T, Influenza = T)}{P(Fever = T)} = \frac{0.669}{0.7479} = 0.8916967509$ or 89.17%.

### Part 2

$P(Influenza = F | Fever = T) = \frac{P(Fever = T | Influenza = F) * P(Influenza = F)}{P(Fever = T)}$

The top of the equation can be reduced as such: $P(Fever = T | Influenza = F) * P(Influenza = F) = P(Fever = T, Influenza = F).$

0.5 * 0.2 * 0.1 * 0.0 +

Also, $P(Fever = T) = \sum_{i}\sum_{j}\sum_{k}P(W = i) * P(C = j | W = i) * P(I = k | W = i) * P(F = T | C = j, I = k) =$
$0.5 * 0.8 * 0.5 * 0.0 +
0.5 * 0.8 * 0.5 * 0.9 +

Finally, $P(Influenza = F | Fever = T) = \frac{P(Fever = T | Influenza = F) * P(Influenza = F)}{P(Fever = T)} = \frac{P(Fever = T, Influenza = F)}{P(Fever = T)} = \frac{0.081}{0.7479} = 0.10830324909$ or 10.8303%.

Another way to check is that $1 - P(Infleunza = T | Fever = T) = 1 - 0.8916967509 = 0.1083032491 = P(Influenza = F | Fever = T).$

Using Pyagrum, after we created a Bayesian Network that was exactly the same as the one displayed in the lab document, set the evidence as Fever: 'True', and then queried for the probabilities for all of the states of Influenza, we were able to obtain a data structure that contains Probabilities for the two states (True and False) of Influenza. All of the code in this exercise was taken from Lab08-BayesianNetwork 2022W.ipynb on D2L.

```python
import os, random
import matplotlib.pyplot as plt
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
import numpy as np

%matplotlib inline

r_seed = 101

# for reproducibility
# must to run this before create the BN! 
# Even in the cases where were changed the states names
gum.initRandom(r_seed) 
random.seed(r_seed)

# fast create a BN (random paramaters are chosen)
bn1 = gum.fastBN('Winter->Influenza;' + 
                'Winter->Cold;' +
                'Influenza->Fever<-Cold')

bn1.cpt('Winter')[:] = [0.5, 0.5]

bn1.cpt('Cold')[{'Winter':0}] = [0.8, 0.2]
bn1.cpt('Cold')[{'Winter':1}] = [0.2, 0.8]

bn1.cpt('Influenza')[{'Winter':0}] = [0.5, 0.5]
bn1.cpt('Influenza')[{'Winter':1}] = [0.1, 0.9] # should be [0.1, 0.9]

bn1.cpt('Fever')[{'Influenza':0, 'Cold':0}] = [1.0, 0.0]
bn1.cpt('Fever')[{'Influenza':0, 'Cold':1}] = [0.1, 0.9]
bn1.cpt('Fever')[{'Influenza':1, 'Cold':0}] = [0.1, 0.9]
bn1.cpt('Fever')[{'Influenza':1, 'Cold':1}] = [0.01, 0.99]

# evidence of Fever is True (1)
evidence = {'Fever': True}

ie = gum.LazyPropagation(bn1)

ie.setEvidence(evidence)
ie.makeInference()

res_influenza = ie.posterior('Influenza')

# calculating the probability that somebody has Influenza considering that he/she already has Fever
print('P(Influenza = True | Fever = True): %.4f%%' % (res_influenza[1] * 100))
print('P(Influenza = False | Fever = True): %.4f%%' % (res_influenza[0] * 100))
```

```python
import os, random
import matplotlib.pyplot as plt
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
import numpy as np

%matplotlib inline

r_seed = 101

# for reproducibility
# must to run this before create the BN! 
# Even in the cases where were changed the states names
gum.initRandom(r_seed) 
random.seed(r_seed)

# fast create a BN (random paramaters are chosen)
bn1 = gum.fastBN('Winter->Influenza;' + 
                'Winter->Cold;' +
                'Influenza->Fever<-Cold')

bn1.cpt('Winter')[:] = [0.5, 0.5]

bn1.cpt('Cold')[{'Winter':0}] = [0.8, 0.2]
bn1.cpt('Cold')[{'Winter':1}] = [0.2, 0.8]

bn1.cpt('Influenza')[{'Winter':0}] = [0.5, 0.5]
bn1.cpt('Influenza')[{'Winter':1}] = [0.1, 0.9] # should be [0.1, 0.9]

bn1.cpt('Fever')[{'Influenza':0, 'Cold':0}] = [1.0, 0.0]
bn1.cpt('Fever')[{'Influenza':0, 'Cold':1}] = [0.1, 0.9]
bn1.cpt('Fever')[{'Influenza':1, 'Cold':0}] = [0.1, 0.9]
bn1.cpt('Fever')[{'Influenza':1, 'Cold':1}] = [0.01, 0.99]

# Calculating the joint probability: Chain Rule
pWCF = bn1.cpt('Fever') * bn1.cpt('Cold') * bn1.cpt('Winter') * bn1.cpt('Influenza')

# the result are all the possible scenarios
pWCF

# In cases where Influenza is "unknown" we can marginalize that variable
pWCF = pWCF.margSumOut(['Influenza'])
pWCF.normalize()

print('P(Fever = F, Cold = F, Winter = F): %.4f%%' % (pWCF[0,0,0] * 100))
print('P(Fever = F, Cold = F, Winter = T): %.4f%%' % (pWCF[0,0,1] * 100))
print('P(Fever = F, Cold = T, Winter = F): %.4f%%' % (pWCF[0,1,0] * 100))
print('P(Fever = F, Cold = T, Winter = T): %.4f%%' % (pWCF[0,1,1] * 100))
print('P(Fever = T, Cold = F, Winter = F): %.4f%%' % (pWCF[1,0,0] * 100))
print('P(Fever = T, Cold = F, Winter = T): %.4f%%' % (pWCF[1,0,1] * 100))
print('P(Fever = T, Cold = T, Winter = F): %.4f%%' % (pWCF[1,1,0] * 100))
print('P(Fever = T, Cold = T, Winter = T): %.4f%%' % (pWCF[1,1,1] * 100))
```

Therefore, there should be 20 entries defined in the Susceptibility CPT (where each of the 20 entries has its own P(S = low), P(S = medium), and P(S = high)). 

In terms of PyAgrum, having an incomplete CPT means that PyAgrum randomly creates values for the missing entries. As these values are randomly created, any Conditional or Joint Probabilities that required the randomly generated Susceptibility entries will vary between code runs. Therefore, it is not possible to gain a particular probability value since PyAgrum is randomly creating them anyway. On the other hand, if we are talking about manually calculating probabilities, then it becomes impossible to calculate certain probabilities. We would either have to not consider the entire data set in order to calculate the probabilities or just not be able to calculate them at all. For example, if we consider P(Gender = Female, Susceptibility = medium), then we can use the Chain Rule and Total Probability to obtain: 

$P(Gender = Female, Susceptibility = medium) = P(Gender = Female) * P(Age) * P(Susceptibility = medium | Age, Gender = Female).$

This becomes:
$\sum{i}P(Gender = Female) * P(Age = i) * P(Susceptibility = medium | Age = i, Gender = Female).$

Therefore, in order to be able to compute P(Gender = Female, Susceptibility = medium), we would require the probabilities/percentages of Females at every Age bracket that have medium Susceptibility to COVID. As we only have the Susceptibilities for Females with Ages 0-9, 40-49, and 90-99, it becomes impossible to calculate the original Probability that we wanted to find (as our CPT table is incomplete). This is obviously due to the Age bins we consider when we are manually calculating the probabilities compared to when PyAgrum calculates them. When we manually calculate the probabilities, we only consider the Age bins that we actually have Susceptibility data for which are the Age bins 0 (for ages 0-9), 4 (for ages 40-49), and 9 (for ages 90-99). On the other hand, when we create our Age CPT, we consider all Age bins which means that our Susceptibility table also requires an entry for every Age bin. However, as we only input Susceptibility data for the Age bins 0, 4, and 9, our Bayesian network creates data for all of the other Susceptibility entries. Therefore, the Bayesian network considers all of the Age bins when performing its calculations and this results in probabilities that are vastly different than the ones we calculated. For example, the manually calculated result and the PyAgrum result for P(Age = 40-49 | Gender = Male, Susceptibility = High) = 0.9388 and P(Age = 40-49 | Gender = Male, Susceptibility = High) = 0.1539 respectively. Also, the manually calculated result and the PyAgrum result for P(Age = 90-99 | Gender = Male, Susceptibility = High) = 0.0612 and P(Age = 90-99 | Gender = Male, Susceptibility = High) = 0.010 respectively. As can be seen there is a difference in probability of 0.78 in the first case and a difference of 0.05 in the second case. The only probabilities that were calculated the same manually and by PyAgrum are the probabilities for P(Age = 0-9 | Gender = Male, Susceptibility = High). This is because the Susceptibility for an individual who is Male and has an Age of 0-9 has a Susceptibility of 0.00. If we look at the manual calculations, then the top half of the Bayes' formula will always equate to 0 regardless of the number of Age bins we are considering. The number of Age bins only affects the bottom of the equation in this case, and so both the manual calculation and PyAgrum perform the step of dividing 0 by a number which results in 0. As a result, both of them produce the same probability for P(Age = 0-9 | Gender = Male, Susceptibility = High). Also, we found that an individual is most likely to be part of the 40-49 Age group given the evidence that the individual is Male and has High Susceptibility (with a probability of 93.88%). However, PyAgrum found that an individual that is Male and has High Susceptibility is most likely to be a part of the 70-79 Age group (with a probability of 29.03%). Once again there is a significant difference in the manual calculation and the PyAgrum calculation because we do not consider the entirety of the Age bins in the manual calculations and PyAgrum creates its own data becuase it is considering every Age bin. Overall, as we have seen before, PyAgrum offers a fast and easy way to make inferences and calculate probabilities when compared to manual calculations. A problem however, is that the Bayesian network used when using PyAgrum needs to be fully filled out otherwise it may produce variable results that are not an accurate representation of a situation.

```python
# male, female
gender = [55, 45]

# stratified data from JIID's report, 2020: https://www.niid.go.jp/niid/en/2019-ncov-e/9417-covid-dp-fe-02.html
age = [16, 23, 347, 428, 334, 398, 923, 1015, 216, 11]
age_labels = ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-89','90-99']

bn = gum.fastBN('Age{0-9|10-19|20-29|30-39|40-49|50-59|60-69|70-79|80-89|90-99}->Susceptibility{low|medium|high};' +
                'Gender{male|female}->Susceptibility;')

bn.cpt('Age')[:] = age / np.sum(age)
bn.cpt('Gender')[:] = gender / np.sum(gender)

# 0-9 yo, male: Susceptibility is 95% "low"
bn.cpt('Susceptibility')[{'Age':0, 'Gender':0}] = [.95, .05, .00]

# 0-9 yo, female: Susceptibility is 99% "low"
bn.cpt('Susceptibility')[{'Age':0, 'Gender':1}] = [.99, .01, .00]

# 90-99 yo, male: Susceptibility is 99% "high"
bn.cpt('Susceptibility')[{'Age':9, 'Gender':0}] = [.00, .01, .99]

# 90-99 yo, female: Susceptibility is 98% "high"
bn.cpt('Susceptibility')[{'Age':9, 'Gender':1}] = [.00, .02, .98]

# 40-49 yo, male: Susceptibility is 50% "high"
bn.cpt('Susceptibility')[{'Age':4, 'Gender':0}] = [.20, .30, .50]

# 40-49 yo, female: Susceptibility is 40% "medium"
bn.cpt('Susceptibility')[{'Age':4, 'Gender':1}] = [.30, .40, .30]

evidence = {'Gender':0, 'Susceptibility':2}

ie = gum.LazyPropagation(bn)

ie.setEvidence(evidence)
ie.makeInference()

gnb.showInference(bn, size='10', evs=evidence)

res_age = ie.posterior('Age')

res_age_a = res_age.toarray()

max_idx = np.argmax(res_age_a)
max_perc = res_age_a[max_idx]*100
max_age = age_labels[max_idx]

print('Males with age [%s] are the most likely with %.2f%% to have HIGH susceptibility' % (max_age, max_perc))
print('Males with age [%s] are likely with %.2f%% to have HIGH susceptibility' % (age_labels[0], res_age_a[0]*100))
print('Males with age [%s] are likely with %.2f%% to have HIGH susceptibility' % (age_labels[4], res_age_a[4]*100))
print('Males with age [%s] are likely with %.2f%% to have HIGH susceptibility' % (age_labels[9], res_age_a[9]*100))
```
